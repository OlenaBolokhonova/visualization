```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Setting a seed for reproducibility
np.random.seed(613)

# Generating data
x = np.arange(50)
y1 = np.random.randint(0,100,50)
y2 = np.random.randint(0,100,50)

# Plotting y1, y2 variables on the same axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1)
ax.plot(x,y2)
fig.show()
```

```python
# Plotting a line plot on the axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1") #add label to y1
ax.plot(x,y2, label = "Person 2") #add label to y2
ax.legend(loc = "lower right") #add legend
fig.show()
```

```python
# Plotting a line plot on the axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1")
ax.plot(x,y2, label = "Person 2") 
ax.legend(loc = "lower right",
          frameon = True, #New addition
          fontsize = 12, #New addition
          ncol = 2, #New addition
          shadow = True) #New addition
fig.show()
```

```python
# Plotting a line plot on the axes
fig, ax = plt.subplots(figsize=(5, 3))
ax.plot(x,y1, label = "Person 1") 
ax.plot(x,y2, label = "Person 2") 
ax.legend(loc = 'upper left', #New addition
          bbox_to_anchor=(1,1)) #New addition
fig.show()
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')
ax.text(10,95, "This value is important!") #Add text function
fig.show()
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')
ax.text(10,95, "This value is important!",
        ha='center', #alignment of text
        color='red', #modify font color
        size = 20) #modify font size
fig.show()
```

```python
# Add empty set of axes
fig, ax = plt.subplots()
ax.axis([0,10,0,10])

#Add positioning text annotation with refernce to the data
ax.text(1,5, " Data: (1,5)", transform=ax.transData )
ax.text(0.5,0.1, " Axes: (0.5,0.1)", transform=ax.transAxes)
ax.text(0.2,0.2, " Figure: (0.2,0.2)", transform=fig.transFigure)
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')

#Add the annotate() function to annotate with arrows ant text (e.g. to focus at a specific point)
ax.annotate('This is important!', xy=(10,95), xytext=(20,94),
            arrowprops=dict(arrowstyle = "wedge", #customize arrow
                            color = 'hotpink')) #customize arrow
plt.show()
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')

# Removing tick marks or labels from axis
ax.yaxis.set_major_locator(plt.NullLocator()) #remove ticks and labels
ax.xaxis.set_major_formatter(plt.NullFormatter()) #remove only labels
plt.show()
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')

#Defining a maximum number of ticks on axis
ax.xaxis.set_major_locator(plt.MaxNLocator(3))
plt.show()
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')

# Set tick marks at a particular interval using MultipleLocator
ax.xaxis.set_major_locator(plt.MultipleLocator(5))
plt.show()
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')

# Rotate axis labels and horizontal alignment, use xticks()
plt.xticks(rotation=45, ha='right')
plt.show()
```

```python
# Plotting a scatter plot
fig, ax = plt.subplots(figsize=(5, 3))
ax.scatter(x,y1, label = "Person 1") 
ax.scatter(x,y2, label = "Person 2") 
ax.legend(loc = 'lower right')

# Setting the x-axis title with a specific font style and color
plt.xlabel("Shiny New X Axis!", fontdict={'fontname': 'serif', 'fontsize': 12, 'color': 'indigo'})

# Show plot
plt.show()
```
